```python
# Function to print the heading line only once

def checkHeader():
    file=open("Batting_Statistics.csv","r")
    header = file.readline()
    if "Name" not in header:
        file=open("Batting_Statistics.csv","w")
        file.write("Name"+"\t"+"Innings"+"\t"+"Runs"+"\t"+"Balls Faced"+"\t"+"Average"+"\t"+"Boundaries"+"\t"+"Half Centuries"+"\t"+"Centuries"+"\t"+"Ducks"+"\t"+"Highest Score"+"\t"
            +"Lowest Score"+"\t"+"Strike Rate"+"\t"+"Hard Hitting Ability"+"\n")
        file.close()

checkHeader()

# Function to calculate batting statistics of a player

def batting():
    name = input('Enter a player name: ')
    n = int(input("Enter number of innings: "))
    bat = []
    ball = []

    for i in range(n):                                               #loop to insert runs scored and balls faced in each innings into a list
        run = int(input("Enter the runs in each innings: "))
        bat.append(run)
        ball_faced = int(input("Enter number of ball faced: "))
        ball.append(ball_faced)

    boundary = int(input("Enter total number of boundaries: "))
    tot_run = 0
    tot_ball = 0
    count_fifty = 0
    count_hundred = 0
    count_duck = 0

    for i in bat:                                                    #loop to calculate total runs , half-centuries , centuries and ducks
        tot_run = tot_run + i
        if i >= 50 and i < 100:
            count_fifty = count_fifty + 1
        if i >= 100:
            count_hundred = count_hundred + 1
        if i == 0:
            count_duck = count_duck + 1

    for j in ball:                                                   #loop to calculate total balls
        tot_ball = tot_ball+j

    print(name.upper())
    print("Runs in" , n , "innings: " , bat)
    print("Balls faced in" , n , "innings: " , ball)
    print("Total runs scored: " , tot_run)
    print("Total balls faced: " , tot_ball)

    avg_run = round(tot_run/n , 2)                                     #calculate average runs , highest score and lowest score
    print("Batting average: " , avg_run)
    high = max(bat)
    print("Highest score: " , high)
    low = min(bat)
    print("Lowest score: " , low)

    print("Number of ducks: " , count_duck)
    print("Number of half centuries: " , count_fifty)
    print("Number of centuries: " , count_hundred)

    strike_rate = round((tot_run/tot_ball)*100,3)                               #calculate strike rate and hard hitting ability
    print("Strike rate: ", strike_rate)
    hard_hitting_ability = round(boundary/tot_ball,2)
    print("Hard Hitting ability: " , hard_hitting_ability)
    
    file = open("Batting_Statistics.csv" , "a+")
    for row in file:
        file.write("{:<10}, {:<10}, {:<10}, {:<10}, {:<10}, {:<10}, {:<10}, {:<10}, {:<10}, {:<10}, {:<10}, {:<10}\n".format(*row[0]))
    file.write(name+"\t"+str(n)+"\t"+str(tot_run)+"\t"+str(tot_ball)+"\t"+str(avg_run)+"\t"+str(boundary)+"\t"+str(count_fifty)+"\t"+str(count_hundred)+"\t"+str(count_duck)+"\t"+str(high)+"\t"
           +str(low)+"\t"+str(strike_rate)+"\t"+str(hard_hitting_ability) + "\n")
    file.close()
```

```python
def checkHeader():
    file = open("Balling_Statistics.csv","r")
    header = file.readline()
    
    if "Name" not in header:
        file = open("Balling_Statistics.csv","w")
        file.write("Name"+"\t"+"Matches"+"\t"+"Overs"+"\t"+"Maiden"+"\t"+"Runs Conceded"+"\t"+"Total Wickets"+"\t"+"Average"+"\t"+"Economy Rate"+"\t"
                   +"Strike Rate"+"\t"+"Runs per over"+"\t"+"Maximum Wickets"+"\t"+"Minimum Wickets"+"\t"+"\n")
        file.close()

checkHeader()

def balling():
    wicket_list = []
    name = input('Enter a player name: ')
    n = int(input("Enter number of matches: "))

    for i in range(n):                                                            #entering wickets in each match into a list
        wicket_per_match = int(input("Enter the number of wicket in this match: "))
        wicket_list.append(wicket_per_match)

    over = int(input("Enter number of overs: "))
    maiden = int(input("Enter number of maidens: "))
    runs_conceded = int(input("Enter the total runs conceded: "))
    wickets = 0

    for i in wicket_list:                                                         #calculate total wickets
        wickets = wickets+i
    
    print("Wickets in",n,"matches: " , wicket_list)
    print("Number of overs: " , over)
    print("Number of maidens bowled: " , maiden)
    print("Total runs conceded: " , runs_conceded)
    print("Total wickets taken: " , wickets)


    bowl_average = runs_conceded / wickets                                        #calculate bowling average
    print("Bowling average:" , bowl_average)
    if bowl_average < 20:
        print(name,"is an extra ordinary bowler")
    elif bowl_average >= 20 and bowl_average < 25:
        print(name,"is an excellent bowler")
    elif bowl_average >= 25 and bowl_average < 30:
        print(name,"is a good bowler")
    elif bowl_average >= 30 and bowl_average < 35:
        print(name,"is an average bowler")
    elif bowl_average >=35 and bowl_average < 40:
        print(name,"is a below average bowler")
    else:
        print(name,"has minimal bowling skills")
    
    economy_rate = (runs_conceded / (over*6))*6                                   #calculate economy rate , strike rate , runs per over , maximum wickets and minimum wickets
    strike_rate = (over*6) / wickets
    runs_per_over = runs_conceded / over
    max_wicket = max(wicket_list)
    min_wicket = min(wicket_list)

    print("Economy rate: " , economy_rate)
    print("Strike rate: " , strike_rate)
    print("Runs conceded per over: " , runs_per_over)
    print("Maximum number of wickets in a match: " , max_wicket)
    print("Minimum number of wickets in a match: " , min_wicket)


    file = open("Balling_Statistics.csv","a+")
    for row in file:
        file.write("{:<10}, {:<10}, {:<10}, {:<10}, {:<10}, {:<10}, {:<10}, {:<10}, {:<10}, {:<10}, {:<10}, {:<10}\n".format(*row[0]))
    file.write(name+"\t"+str(n)+"\t"+str(over)+"\t"+str(maiden)+"\t"+str(runs_conceded)+"\t"+str(wickets)+"\t"+str(bowl_average)+"\t"+str(economy_rate)+"\t"+str(strike_rate)+"\t"+str(runs_per_over)+"\t"
    +str(max_wicket)+"\t"+"\t"+str(min_wicket)+"\n")
    file.close()

```
